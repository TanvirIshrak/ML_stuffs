# Scatterplot with Plotly

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

student = pd.read_csv('student_dataset_complete.csv')
student
```

```python
fig = px.scatter(student , x='time_study', y ='Marks',
                color='gender' , hover_data=['time_study','Marks','hostel'])
fig
```

# Lineplot with plotly

```python
enrollment = pd.read_csv('enrollment_data.csv')

fig=px.line(enrollment , x='Year' , y='Programming',markers=True)
fig.show()
```

```python
# this will make the graph in a html file in the current directory
fig.write_html('digital_marketing_data.html')
```

# Histogram using plotly

```python
student = pd.read_csv('student_dataset_complete.csv')
student_marks=student[['Marks','study_hours','attendance_rate','gender']]
student_marks
```

```python
fig = px.histogram(student_marks ,  x='attendance_rate')
fig.show()
```

```python
fig = px.histogram(student_marks ,  x='attendance_rate',color='gender',nbins=10)
fig.show()
```
