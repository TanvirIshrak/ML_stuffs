```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

student = pd.read_csv('sns_data.csv')
student
```

```python
sns.scatterplot(data=student, x='study_hours', y='test_score')
# got positive corellation
```

```python
sns.scatterplot(data=student, x='study_hours', y='test_score')
# got positive corellation
```

```python
sns.scatterplot(data=student, x='study_hours', y='test_score',hue='gender', style='subject',size='Tshirt_size')
# got positive corellation
```

```python
# relplot has access to whole graph
sns.relplot(kind='scatter',data=student, x='study_hours', y='test_score',hue='gender', style='subject',size='Tshirt_size')
# got positive corellation
```
