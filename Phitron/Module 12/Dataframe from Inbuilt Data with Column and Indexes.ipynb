```python
import pandas as pd

#list type
list = [["ishrak",22],["fatin",21],["leon",19]]
var = pd.DataFrame(list,columns =['Name','Age'],index=[1,2,3])    # also naming the columns and reindexing

print(var)
print(type(var))
```

```python
# tuple type
tup = (('ishrak',23),("fatin",21),("leon",19))
tup_df = pd.DataFrame(tup,columns =['Name','Age'],index=[1,2,3])
tup_df
```
