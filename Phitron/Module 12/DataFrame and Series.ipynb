## Different file types in Pandas

```python
# CSV file

csv_data = pd.read_csv('student_data.csv')
print(csv_data)
```

```python
# excel file
# for this we need to install "openpyxl" from terminal
import pandas as pd
excel_data = pd.read_excel('phitron_student_marks.xlsx')
print(excel_data)
```

```python
# parquet file

import pandas as pd
pk_file = pd.read_parquet('students.parquet')
print(pk_file)
```

```python
# json file
json = pd.read_json('data.json')
print(json)
print(type(json))
```
