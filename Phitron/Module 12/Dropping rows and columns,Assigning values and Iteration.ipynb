```python
import pandas as pd
df = pd.read_csv('student_data.csv')
df
```

```python
# removing all value
#ROW WISE
df.drop(0) # ( index)
```

```python
# removing all value
# COLUMN WISE
df.drop('Instructor', axis = 1) # ( index)
```

```python
# Assigning value

df.loc[1,'Python Marks'] = 90  # [ row_index , "column name']
df.loc[1,'CompletionStatus'] = 'Completed'  
df.head(5)
```

```python
# Assigning value

df.loc[1,'Python Marks'] += 2  # [ row_index , "column name']
df.loc[1,'CompletionStatus'] = 'Completed'  
df.head(5)
```

```python
# Iterateing values
for i,series in df.iterrows():
    print(f"{i} : {series}")
    print("\n")
```
