```python
import pandas as pd

df = pd.read_csv("student_data.csv")
df
```

```python
# this removes NuN value contained rows
df.dropna()
```

```python
# removes if all values are null in a row
df.dropna(how = 'all',inplace = True)
df
```

```python
# removing NaN values of a dedicated column
df.dropna(subset = ['Python Marks'])
```

```python
# filling the null values with something else
df.fillna(0)
```
