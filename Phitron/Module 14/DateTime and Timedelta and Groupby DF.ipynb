```python
import pandas as pd
df = pd.read_csv('student_completed_data.csv')
df
```

```python
df['EnrollmentDate']
```

```python
df['EnrollmentDate'] = pd.to_datetime(df['EnrollmentDate'])
df['EnrollmentDate']
```

```python
df['Enrollment Year'] = df['EnrollmentDate'].dt.year
df
```

```python
df['Enrollment Year'] = df['EnrollmentDate'].dt.day
df
```

```python
df['FinishedDate'] = pd.to_datetime(df['FinishedDate'])
df['FinishedDate']
```

```python
df['Total time to finish'] = df['FinishedDate'] - df['EnrollmentDate']
df
```
