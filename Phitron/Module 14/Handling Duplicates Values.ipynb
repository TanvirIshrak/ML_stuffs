```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "x", "Frank", "Grace", "Hannah","Sakib"],
    "City": ["New York", "Los Angeles", "Newark", "New York", "New Delhi", "Chicago", "New Orleans", "Houston","H Los Ang"],
    "Department": ["HR", "IT", "Finance", "HR", "x", "Marketing", "Finance", "HR", "HR"],
    "Salary": [50000, 60000, 55000, 50000, 50000, 58000, 62000, 51000,70000]
}

df = pd.DataFrame(data)
df
```

```python
df.duplicated().sum()
```

```python
# deleting duplicates values

df.drop_duplicates()  # deletes the duplivated one
df
```

```python
df.drop_duplicates(inplace = True)  # deletes the duplivated one
df
```

```python
df
```

# deleting column wise

```python
df.drop_duplicates(subset=['Salary'], keep='last')
# in case of keep='last' means it would delete the first one
```
