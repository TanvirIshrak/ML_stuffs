```python
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

#### Answer to the Question no 1

```python
age_arr = np.array(ages)
salary_arr = np.array(salaries)

print("ages :",age_arr)
print("salaries :",salary_arr)

print("ages dtype: ",age_arr.dtype)
print("ages ndim: ",age_arr.ndim)
print("ages shape: ",age_arr.shape)
print("ages size: ",age_arr.size)
print("salary dtype: ",salary_arr.dtype)
print("salary ndim: ",salary_arr.ndim)
print("salary shape: ",salary_arr.shape)
print("salary size: ",salary_arr.size)
```

#### Answer to the Question no 2

```python
max_sal = np.max(salary_arr)
min_sal = np.min(salary_arr)
avg_sal = np.mean(salary_arr)
avg_age = np.mean(age_arr)

print("Max salary: ",max_sal)
print("Min salary: ",min_sal)
print("Avg sal: ", avg_sal)
print("Avg age: ", avg_age)
```

#### Answer to the Question no 3

```python
older_30 = age_arr[age_arr > 30]
count = older_30.size

print("older than 30: ",older_30)
print("Count of them: ",count)
```

#### Answer to the Question no 4

```python
new_ages = age_arr + 5
print("Original: ",age_arr)
print("New: ",new_ages)
```

#### Answer to the Question no 5

```python
total_sal = np.sum(salary_arr)
salary_range = np.max(salary_arr) - np.min(salary_arr)

print("Total salary expense: ",total_sal)
print("Difference: ",salary_range)
```
