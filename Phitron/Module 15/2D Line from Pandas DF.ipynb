```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('enrollment_data.csv')
df
```

```python
plt.xlabel('Years')
plt.ylabel('Count of enrollment')
plt.title('Student entollment over 8 year')

plt.plot(df['Year'] , df['Programming'])
plt.plot(df['Year'] , df['Digital Marketing'])
```

```python
# adding labels using plt.legent()

plt.xlabel('Years')
plt.ylabel('Count of enrollment')
plt.title('Student entollment over 8 year')

plt.plot(df['Year'] , df['Programming'], label='Programming')
plt.plot(df['Year'] , df['Digital Marketing'], label='Digital Marketing')

# we can relocate the functiin plt.legend()
plt.legend()
# plt.legend(loc='lower right')
```

# Customizing Line plot ( color,linestyle,marker) and Grid

```python
# adding labels using plt.legent()

plt.xlabel('Years')
plt.ylabel('Count of enrollment')
plt.title('Student entollment over 8 year')

# adding label , changing colour , customising linewidth and adding marker
plt.plot(df['Year'] , df['Programming'], label='Programming', color='green', linewidth=5, marker='x')
plt.plot(df['Year'] , df['Digital Marketing'], label='Digital Marketing',color='red',linestyle='dashed',marker='o')


plt.legend()
plt.grid()
# plt.show()
```
