```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('student_data.csv')
data
```

```python
# locationwise
location = data.groupby('Location').size()
                                        # auto percentage
plt.pie(location,labels=location.index , autopct = '%1.1f%%')
plt.show()
```

```python
# locationwise to show any part particularly
location = data.groupby('Location').size()
                                        # auto percentage
plt.pie(location,labels=location.index , autopct = '%1.1f%%' , explode = (0.1,0,0,0,0))
plt.show()
```
