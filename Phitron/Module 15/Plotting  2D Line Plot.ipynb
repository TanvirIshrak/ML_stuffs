```python
# 2D line plot
# -> how a value changes over time
# -> time data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

hour = [2,3,4,1,5,7,3]
days=[1,2,3,4,5,6,7]

plt.plot(days, hour)  #( x-axis er value , y-axis er value)
```

```python
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Trend of a student study time over a week')

plt.plot(days, hour)
```
