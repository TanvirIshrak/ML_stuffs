```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
df
```

```python
# genderwise

gender = df.groupby('Gender').size()
gender.index
# gender.values
```

```python
plt.bar(gender.index , gender.values , color=['red','blue'])
plt.xlabel('Genders')
plt.ylabel('Count')
```

```python
data = pd.read_csv('student_data.csv')
data
```

```python
status = data.groupby('CompletionStatus').size()
status
```

```python
plt.bar(status.index , status.values , color=['orange','blue','purple'])
plt.xlabel('CompletionStatus')
plt.ylabel('Count')
```

```python
# locationwise
location = data.groupby('Location').size()
plt.bar(location.index , location.values , color=['orange','blue','purple'])
plt.xlabel('CompletionStatus')
plt.ylabel('Count')
```
