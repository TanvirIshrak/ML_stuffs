```python
# how a data distributed over a factor
# -> frequency

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

students = pd.read_csv('student_IQdata.csv')
students
```

```python
plt.xlabel('IQ Score')
plt.ylabel('Frequency of students')

plt.hist(students['IQ_Score'], color='Red' , edgecolor='black')
```

```python
plt.xlabel('Study_Hour')
plt.ylabel('Frequency of students')

plt.hist(students['Study_Hour'], bins=20,color='Red' , edgecolor='black')
```

```python
plt.xlabel('Study_Hour')
plt.ylabel('Frequency of students')

plt.hist(students['Study_Hour'], bins=[1,2,3,4,5,6,7,8,9],color='Red' , edgecolor='black')
                        # means 1 theke 2 , 2 theke 3 and proceed
```
