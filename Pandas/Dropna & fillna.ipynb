#  Dropna & fillna

```python
import pandas as pd

csv_1=pd.read_csv("D:\\Python learning sequence\\python for ML\\Pandas\\drop_fill copy.csv")
csv_1
```

```python
# this function will remove all the NuN component having rows
csv_1.dropna()  # by default it's action axis is 0 means row
```

```python
# this function will remove all the NuN component having columns
#csv_1.dropna(axis=1)  
```

```python
csv_1.dropna(how="all")   #this parameter removes the row having null values in all columns only
#csv_1.dropna(how="any")  #this will remove all the null cell having rows
```

```python
# to remove null values from a perticular column 
csv_1.dropna(subset=["Job Title"])
```

# Fillna

```python
# to fill all the null cell with something else
csv_1.fillna("python xs")
```

```python
# filling specific columns with different values
csv_1.fillna({"First Name":"hexasol", "Sex":"C"})
```

```python
# to fill the null value with it's previous or next value
#csv_1.fillna(method="ffill")  #forward feeling
csv_1.fillna(method="bfill")  #backward feeling
```
